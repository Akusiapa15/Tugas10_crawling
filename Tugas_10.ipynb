```python
import requests
from csv import writer
import json
```

```python
url = "https://catalog-api5.sociolla.com/search?limit=16&skip=0&sort=-created_at&filter=%7B%22categories.id%22:%225d3d502437382b158ee6dda5%22%7D"
response = requests.get(url)
response = response.json()
list_item = response['data']
for element in list_item:
  print(element)
```

```python
with open('/content/sample_data/eyeshadow.csv', 'w') as csv_file:
    csv_writer = writer(csv_file)
    headers = ['eyeshadow']
    csv_writer.writerow(headers)

    for item in list_item:
        # Convert the dictionary to a string before replacing newline characters
        item_str = str(item)
        item_str = item_str.replace("\n", "")
        csv_writer.writerow([item_str]) # Write the modified string to the CSV
```

```python
with open('/content/sample_data/nweyeshadow.csv', 'w') as csv_file:
    csv_writer = writer(csv_file)
    headers = ['eyeshadow']
    csv_writer.writerow(headers)

    for item in list_item:
        # Create the string using f-string formatting for easier readability and flexibility.
        item_str = f"{item.get('_id', '')},{item.get('name', '')},{item.get('region', '')},{item.get('min_price', '')},{item.get('max_price', '')},{item.get('description', '')}"
        item_str = item_str.replace("\n", "")
        csv_writer.writerow([item_str]) # Write the modified string to the CSV
```
